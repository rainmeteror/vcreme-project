Triple Exponential Moving Average: 30
- EMA_1 = Exponential Moving Average (EMA)
- EMA_2 = EMA of EMA_1
- EMA_3 = EMA of EMA_2

- TEMA = 3*(EMA_1) - 3*(EMA_2) + EMA_3

```python
vcb = df.loc[df['Ticker'] == 'VCB']
vcb
```

```python
vcb['ema_1'] = vcb['Close'].ewm(span=30, adjust=False).mean()
vcb['ema_2'] = vcb['ema_1'].ewm(span=30, adjust=False).mean()
vcb['ema_3'] = vcb['ema_2'].ewm(span=30, adjust=False).mean()

vcb['tema'] = 3*vcb['ema_1'] - 3*vcb['ema_2'] + vcb['ema_3']

vcb
```

```python
df['ema_1'] = df.groupby('Ticker')['Close'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()
df['ema_2'] = df.groupby('Ticker')['ema_1'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()
df['ema_3'] = df.groupby('Ticker')['ema_2'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()

df['tema'] = 3*df['ema_1'] - 3*df['ema_2'] + df['ema_3']
df
```

```python
def triple_ema(panel_data):
    panel_data['ema_1'] = panel_data.groupby('Ticker')['Close'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()
    panel_data['ema_2'] = panel_data.groupby('Ticker')['ema_1'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()
    panel_data['ema_3'] = panel_data.groupby('Ticker')['ema_2'].ewm(span=30, adjust=False).mean().reset_index(drop=True).to_list()

    panel_data['tema'] = 3*panel_data['ema_1'] - 3*panel_data['ema_2'] + panel_data['ema_3']
    
    del panel_data['ema_1'], panel_data['ema_2'], panel_data['ema_3']
    
    return panel_data
```

```python
df = triple_ema(df)
df
```
