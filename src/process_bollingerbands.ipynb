```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"E:\Tung\Python\VcreamProject\cache\ALL.csv")

df.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

Bollinger Bands: MA10, MA20, MA30

```python
df["ma_10"] = df.groupby('Ticker')['Close'].rolling(10, min_periods=1).mean().reset_index(drop=True).to_list()
df["upper_b10"] = df["ma_10"] + df.groupby('Ticker')['Close'].rolling(10, min_periods=1).std().reset_index(drop=True).to_list()
df["lower_b10"] = df["ma_10"] - df.groupby('Ticker')['Close'].rolling(10, min_periods=1).std().reset_index(drop=True).to_list()

df["ma_20"] = df.groupby('Ticker')['Close'].rolling(20, min_periods=1).mean().reset_index(drop=True).to_list()
df["upper_b20"] = df["ma_20"] + df.groupby('Ticker')['Close'].rolling(20, min_periods=1).std().reset_index(drop=True).to_list()
df["lower_b20"] = df["ma_20"] - df.groupby('Ticker')['Close'].rolling(20, min_periods=1).std().reset_index(drop=True).to_list()

df["ma_30"] = df.groupby('Ticker')['Close'].rolling(30, min_periods=1).mean().reset_index(drop=True).to_list()
df["upper_b30"] = df["ma_30"] + df.groupby('Ticker')['Close'].rolling(30, min_periods=1).std().reset_index(drop=True).to_list()
df["lower_b30"] = df["ma_30"] - df.groupby('Ticker')['Close'].rolling(30, min_periods=1).std().reset_index(drop=True).to_list()
```
