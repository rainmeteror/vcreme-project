```python
vcb = df.loc[df['Ticker'] == 'VCB']
vcb
```

```python
vcb['higher_close'] = np.where(vcb['Close'] >= vcb['Close'].shift(1), vcb['Close'].diff(1), 0)
vcb['lower_close'] = np.where(vcb['Close'] < vcb['Close'].shift(1), vcb['Close'].diff(1)*(-1), 0)
vcb['chande_mo'] = (vcb['higher_close'].rolling(14).sum() - vcb['lower_close'].rolling(14).sum()) / \
    (vcb['higher_close'].rolling(14).sum() +
     vcb['lower_close'].rolling(14).sum()) * 100
    
vcb
```

```python
df['previous_close'] = df.groupby('Ticker')['Close'].shift(1)

df['delta'] = df.groupby('Ticker')['Close'].diff(1)
df['higher_close'] = np.where(df['delta']>=0, df['delta'], 0)
df['lower_close'] = np.where(df['delta']<0, df['delta']*(-1), 0)
df['higher_close_14'] = df.groupby('Ticker')['higher_close'].rolling(14).sum().reset_index(drop=True).to_list()
df['lower_close_14'] = df.groupby('Ticker')['lower_close'].rolling(14).sum().reset_index(drop=True).to_list()
df['chande_mo_14'] = (df['higher_close_14'] - df['lower_close_14'])/(df['higher_close_14'] + df['lower_close_14'])*100

df
```

```python
def chandeMO(panel_data, lookback:int):
    panel_data['previous_close'] = panel_data.groupby('Ticker')['Close'].shift(1)
    panel_data['delta'] = panel_data.groupby('Ticker')['Close'].diff(1)
    panel_data['higher_close'] = np.where(panel_data['delta']>=0, panel_data['delta'], 0)
    panel_data['lower_close'] = np.where(panel_data['delta']<0, panel_data['delta']*(-1), 0)
    panel_data['higher_close_14'] = panel_data.groupby('Ticker')['higher_close'].rolling(lookback).sum().reset_index(drop=True).to_list()
    panel_data['lower_close_14'] = panel_data.groupby('Ticker')['lower_close'].rolling(lookback).sum().reset_index(drop=True).to_list()
    panel_data[f'chande_mo_{lookback}'] = (panel_data['higher_close_14'] - panel_data['lower_close_14'])/(panel_data['higher_close_14'] + panel_data['lower_close_14'])*100
    
    del panel_data['previous_close']
    del panel_data['delta']
    del panel_data['higher_close']
    del panel_data['lower_close']
    del panel_data['higher_close_14']
    del panel_data['lower_close_14']
    
    return panel_data
    
```

```python
df = chandeMO(panel_data=df, lookback=14)
df
```
