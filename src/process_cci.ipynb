```python
vcb = df.loc[df['Ticker'] == 'VCB']
vcb
```

```python
vcb['typical_price'] = np.mean(vcb[['High', 'Low', 'Close']], axis=1)
vcb['tp_ma_14'] = vcb['typical_price'].rolling(14).mean()
vcb['mad_ma_14'] = vcb['typical_price'].rolling(14).apply(lambda x: pd.Series(x).mad())
vcb['CCI_14'] = (vcb['typical_price']-vcb['tp_ma_14'])/(0.015*vcb['mad_ma_14'])

vcb
```
